# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

**Plotting Side-by-side**

This is largely accomplished by `plt.subplots()`. Returns a `figure` object along with an iterable of `axes`. If working with 2-3 subplots only, typically better to unpack the axes immediately into distinct variables

```python
x = np.linspace(-1, 1)
y = x ** 2
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
ax1.plot(x, y)
ax2.plot(x, -y);
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1)
ax1.plot(x, y)
ax2.plot(x, -y);
```
